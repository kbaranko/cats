```python
def tuplify(df):
    tup_list = []
    
    for index, row in df.iterrows():
        tup = (row['Country'], 
               row['Regulatory_Quality'], 
               row['Political_Stability'], 
               row['Accountability'], 
               row['Rule_of_Law']
              ) # tuple length 7
        tup_list.append(tup)
    return tup_list
```
